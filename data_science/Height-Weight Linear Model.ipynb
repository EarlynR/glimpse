```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
import cPickle
```

```python
height_weight_df = pd.read_csv('gender-height-weight.csv')
```

```python
height_weight_df.head()
```

```python
del height_weight_df['Height (cm)']
del height_weight_df['Weight (kg)']
```

```python
height_weight_df.head()
```

```python
height_weight_df.groupby('Gender').describe()
```

```python
female_height = height_weight_df.groupby('Gender').mean()['Height']
male_height = height_weight_df.groupby('Gender').mean()['Height'] 
```

```python
#Make dummy variables for Gender.
df_gender = pd.get_dummies(height_weight_df['Gender'])
```

```python
height_weight_gender_df = pd.concat([height_weight_df, df_gender], axis=1)
```

```python
del height_weight_gender_df['Gender']
```

```python
height_weight_gender_df
```

```python
weight = height_weight_gender_df['Weight']
predictors = height_weight_gender_df[['Height','Female', 'Male']]
```

```python
train_set, test_set, train_target, test_target = train_test_split(predictors, weight, test_size=0.33, random_state=42)
```

```python
linear_regression = LinearRegression(fit_intercept= True, normalize= True, n_jobs= -1)
```

```python
linear_regression.fit(train_set, train_target)
```

```python
linear_regression.predict(test_set)
```

```python
linear_regression.score(test_set, test_target)
```

```python
with open('linear_regression_model.pkl', 'wb') as fid:
    cPickle.dump(linear_regression, fid) 
```
